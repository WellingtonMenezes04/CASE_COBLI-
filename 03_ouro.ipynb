```python
df_ouro = spark.read.table("default.tabela_acidentes_sp")

df_ouro.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("default.tabela_acidentes_sp") 
```

```python
%sql
SELECT * FROM tabela_acidentes_sp
```

```python
%sql
--Mostrando o total de acidentes por data

SELECT data, COUNT(*)
FROM tabela_acidentes_sp
GROUP BY data
ORDER BY data
```

```python
%sql
--Mostrando o total de acidentes

SELECT 
 COUNT(*) as total_acidentes
 FROM tabela_acidentes_sp
```

```python
%sql
-- Total de acidentes numa sexta-feira
SELECT dia_semana, COUNT(*) as total_acidentes_na_sexta
FROM tabela_acidentes_sp
WHERE dia_semana = 'sexta-feira'
GROUP BY dia_semana
```
